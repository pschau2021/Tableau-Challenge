```python
data2018to21 = pd.concat([data2018, data2019, data2020, data2021]).reset_index(drop=True)
data2018to21.sample(5)
```

```python
# investigating whether there are suspicious data points
data2018to21['birth year'].unique()
```

```python
# filtering out dobs that would make the user greater than 90 years old
d201821 = data2018to21[data2018to21['birth year'] >= 1931]
print(d201821['birth year'].unique())
d201821.sample(5)
```

```python
# copying index to a column to create a unique id for each trip
d201821['tripID'] = d201821.index
d201821 = d201821[['tripID', 'tripduration', 'starttime', 'stoptime', 'start station id',
    'start station name', 'start station latitude',
    'start station longitude', 'end station id', 'end station name',
    'end station latitude', 'end station longitude', 'bikeid', 'usertype',
    'birth year', 'gender']]
d201821.sample(5)
```

```python
# exporting full dataset
d201821.to_csv('bike_csvs/newdata1821.csv', index=False)
```
